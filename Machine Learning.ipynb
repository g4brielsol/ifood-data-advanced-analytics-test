## Create a Pipeline for all the transformations to be done after the split in the cross validation fase:

```python
def kid_and_teen(kidhome: int, teenhome: int) -> int:
    """ Function to return if a tuple has a Kid and a Teen"""
    if (kidhome > 0) and (teenhome > 0):
        return 1
    else:
        return 0


def customer_pipeline(df):
    # get outliers
    s = df['Year_Birth']
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    iqr_lower = q1 - 1.5 * iqr
    iqr_upper = q3 + 1.5 * iqr

    # remove 'Year_Birth' outliers
    df = df[(s > iqr_lower) & (s < iqr_upper)]

    # get Income outliers
    s = df['Income']
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    iqr_lower = q1 - 1.5 * iqr
    iqr_upper = q3 + 1.5 * iqr
    df[(s < iqr_lower) | (s > iqr_upper)]

    # remove 'Income' outliers and Null values
    df = df[(s > iqr_lower) & (s < iqr_upper)]

    # remove Incomes below 10000
    df = df[df['Income'] > 10000]

    # get NumWebPurchases outliers
    s = df['NumWebPurchases']
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    iqr_lower = q1 - 1.5 * iqr
    iqr_upper = q3 + 1.5 * iqr
    df[(s < iqr_lower) | (s > iqr_upper)]

    # remove 'NumWebPurchases' outliers
    df = df[(s > iqr_lower) & (s < iqr_upper)]

    # transform datetime string into datetime
    # create day, month, year and day of the week columns

    df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
    df['Dt_Customer_Day'] = df['Dt_Customer'].dt.day
    df['Dt_Customer_Month'] = df['Dt_Customer'].dt.month
    df['Dt_Customer_Year'] = df['Dt_Customer'].dt.year

    days = {0:'Mon',1:'Tues',2:'Weds',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}

    df['Dt_Customer_Weekday'] = df['Dt_Customer'].dt.dayofweek.apply(lambda x: days[x])

    # -------------

    # Droping constant columns
    df = df.drop(['Z_CostContact', 'Z_Revenue'], axis=1)

    index_to_drop = df[(df['Marital_Status'] == 'Alone') | 
                       (df['Marital_Status'] == 'Absurd') |
                       (df['Marital_Status'] == 'YOLO')].index.to_list() 

    # df without 'Alone', 'Absurd' and 'YOLO' in the Marital_Status column 
    df = df.drop(index_to_drop)
    df['Kid_and_Teen_Home'] = list(map(kid_and_teen, df['Kidhome'], df['Teenhome']))

    education = {'Basic': 0, '2n Cycle': 1, 'Graduation': 2, 'Master': 3, 'PhD': 4}

    df['Education_Order'] = df['Education'].apply(lambda x: education[x])
    
    df = pd.get_dummies(df, columns=['Marital_Status'])
    
    kbins = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform')

    df['Year_Birth_Bins'] = kbins.fit_transform(df['Year_Birth'].values.reshape(-1, 1))
    df['Income_Bins'] = kbins.fit_transform(df['Income'].values.reshape(-1, 1))
    df['Recency_Bins'] = kbins.fit_transform(df['Recency'].values.reshape(-1, 1))
    df['MntWines_Bins'] = kbins.fit_transform(df['MntWines'].values.reshape(-1, 1))
    df['MntFruits_Bins'] = kbins.fit_transform(df['MntFruits'].values.reshape(-1, 1))
    df['MntMeatProducts_Bins'] = kbins.fit_transform(df['MntMeatProducts'].values.reshape(-1, 1))
    df['MntFishProducts_Bins'] = kbins.fit_transform(df['MntFishProducts'].values.reshape(-1, 1))
    df['MntSweetProducts_Bins'] = kbins.fit_transform(df['MntSweetProducts'].values.reshape(-1, 1))
    df['MntGoldProds_Bins'] = kbins.fit_transform(df['MntGoldProds'].values.reshape(-1, 1))
    df['NumDealsPurchases_Bins'] = kbins.fit_transform(df['NumDealsPurchases'].values.reshape(-1, 1))
    df['NumWebPurchases_Bins'] = kbins.fit_transform(df['NumWebPurchases'].values.reshape(-1, 1))
    df['NumCatalogPurchases_Bins'] = kbins.fit_transform(df['NumCatalogPurchases'].values.reshape(-1, 1))
    df['NumStorePurchases_Bins'] = kbins.fit_transform(df['NumStorePurchases'].values.reshape(-1, 1))
    df['NumWebVisitsMonth_Bins'] = kbins.fit_transform(df['NumWebVisitsMonth'].values.reshape(-1, 1))
    
    robust = RobustScaler(quantile_range = (0.1,0.9))
    scaler = StandardScaler()

    df['Year_Birth_Stand'] = scaler.fit_transform(df['Year_Birth'].values.reshape(-1, 1))
    df['Income_Stand'] = scaler.fit_transform(df['Income'].values.reshape(-1, 1))
    df['Recency_Stand'] = scaler.fit_transform(df['Recency'].values.reshape(-1, 1))
    df['MntWines_Stand'] = robust.fit_transform(df['MntWines'].values.reshape(-1, 1))
    df['MntFruits_Stand'] = robust.fit_transform(df['MntFruits'].values.reshape(-1, 1))
    df['MntMeatProducts_Stand'] = robust.fit_transform(df['MntMeatProducts'].values.reshape(-1, 1))
    df['MntFishProducts_Stand'] = robust.fit_transform(df['MntFishProducts'].values.reshape(-1, 1))
    df['MntSweetProducts_Stand'] = robust.fit_transform(df['MntSweetProducts'].values.reshape(-1, 1))
    df['MntGoldProds_Stand'] = robust.fit_transform(df['MntGoldProds'].values.reshape(-1, 1))
    df['NumDealsPurchases_Stand'] = scaler.fit_transform(df['NumDealsPurchases'].values.reshape(-1, 1))
    df['NumWebPurchases_Stand'] = scaler.fit_transform(df['NumWebPurchases'].values.reshape(-1, 1))
    df['NumCatalogPurchases_Stand'] = scaler.fit_transform(df['NumCatalogPurchases'].values.reshape(-1, 1))
    df['NumStorePurchases_Stand'] = scaler.fit_transform(df['NumStorePurchases'].values.reshape(-1, 1))
    df['NumWebVisitsMonth_Stand'] = scaler.fit_transform(df['NumWebVisitsMonth'].values.reshape(-1, 1))
    
    return df
```
