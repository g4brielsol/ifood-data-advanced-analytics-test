## Checking outliers on the Continue, Discrete Variables and Year_Birth Column:

## It' better to remove outliers so the data can have a normal distribution
## To remove the outliers, the following function was used:
    s = df[column]
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    iqr_lower = q1 - 1.5 * iqr
    iqr_upper = q3 + 1.5 * iqr
    outliers = dict(s[(s < iqr_lower) | (s > iqr_upper)])
    
    Where we get the diference between the 75% and 25% quantile, and the outliers are varibles more or less
    than 50% of the q3 and q1 difference.

## The columns with few outliers were removed for the data to be more normalized. The columns with few outliers are: 
    * Year_Birth, Income, NumWebPurchases, NumWebVisitsMonth

```python
sns.histplot(data=df, x="Year_Birth")
```

```python
# show outliers
s = df['Year_Birth']
q1 = s.quantile(0.25)
q3 = s.quantile(0.75)
iqr = q3 - q1
iqr_lower = q1 - 1.5 * iqr
iqr_upper = q3 + 1.5 * iqr
df[(s < iqr_lower) | (s > iqr_upper)]
```

```python
# remove 'Year_Birth' outliers
df = df[(s > iqr_lower) & (s < iqr_upper)]
```

```python
sns.histplot(data=df, x="Year_Birth")
```

```python
# remove 'Income' outliers and Null values
df = df[(s > iqr_lower) & (s < iqr_upper)]
```

## Transforming Dt_Customer column:

```python
# transform datetime string into datetime
# create day, month, year and day of the week columns

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['Dt_Customer_Day'] = df['Dt_Customer'].dt.day
df['Dt_Customer_Month'] = df['Dt_Customer'].dt.month
df['Dt_Customer_Year'] = df['Dt_Customer'].dt.year

days = {0:'Mon',1:'Tues',2:'Weds',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}

df['Dt_Customer_Weekday'] = df['Dt_Customer'].dt.dayofweek.apply(lambda x: days[x])
```

```python
columns = ['Dt_Customer', 'Dt_Customer_Day', 'Dt_Customer_Month', 'Dt_Customer_Year', 'Dt_Customer_Weekday']

date_df = df[columns]
date_df
```
